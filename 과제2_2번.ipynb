```python
import csv
file = open("age.csv",encoding="cp949")
age = csv.reader(file)
result = []

area = input('지역명 입력 : ')

for i in age:
  if area in i[0]:
    for j in i[15+3:35+3+1]:
      result.append(int(j.replace(',','')))
print(result)

import matplotlib.pyplot as plt
plt.rc('font',family = 'NanumBarunGothic')
plt.title(area+' 인구구조')
plt.bar(range(15,36), result ,color = 'green', label = '인구수')
plt.plot(range(15,36), result ,color = 'red', label = '인구수')
plt.xlabel('나이')
plt.ylabel('인구수')
plt.legend()
plt.show()
```
